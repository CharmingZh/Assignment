# 阈值操作
```
ret, dist = cv2.threshold(src, thresh, maxval, type)
```
- src, 输入图；
- dst，输出图；
- thresh：阈值；
- maxval：像素超过阈值后，被赋予的值；
- type：二值化操作的类型：
    - `cv2.THRESH_BINARY`：超过阈值部分取`maxval`，否则取$0$；
    - `cv2.THRESH_BINARY_INV`：上述反转；
    - `cv2.THRESH_TRUNC`：大于阈值部分设为阈值，否则不变；
    - `cv2.THRESH_TOZERO`：大于阈值部分不变，其余取$0$；
    - `cv2.THRESH_TOZERO_INV`：上述反转；

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

img = cv2.imread('test.png')

def show(img):
    cv2.imshow('test', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
````

````python
img = cv2.imread('test.png')
ret, t1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
ret, t2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
ret, t3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
ret, t4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
ret, t5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['ORING', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, t1, t2, t3, t4, t5]

for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.savefig('Threshold_diff', dpi=300)
````

# 平滑

````python
img = cv2.imread('noise.jpg')
b, g, r = cv2.split(img)
img = cv2.merge((r, g, b))

blur = cv2.blur(img, (3, 3))
gaussian = cv2.GaussianBlur(img, (5, 5), 1)
median = cv2.medianBlur(img, 3)

titles = ['Ori', 'Blur', 'Gaussian', 'Med']
images = [img, blur, gaussian, median]

for i in range(4):
    plt.subplot(1, 4, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.savefig('Blur_diff.png', dpi=400)
````
