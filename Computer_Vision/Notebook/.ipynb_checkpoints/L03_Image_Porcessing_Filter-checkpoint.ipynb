```python
import sys
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline

def readRGB(filename: str):
    # 读入一个 RGB 图像
    blue, green, red = cv2.split(
        cv2.imread(filename)
    )
    return cv2.merge((red, green, blue))

def readGRAY(filename: str):
    # 读入一个单通道灰度图
    return cv2.cvtColor(
        readRGB(filename),
        cv2.COLOR_BGR2GRAY
    )

def show_gray(pic):
    plt.imshow(pic, cmap = 'gray')

def show(pic_1, pic_2=None, pic_3=None, msg="Ori", msg_2="show", msg_3="show"):
    if pic_3 is None:
        if pic_2 is None:
            plt.imshow(pic_1, cmap='gray')
            plt.title(msg)
            plt.xticks([])
            plt.yticks([])
            return
        else:
            plt.subplot(121)
            plt.imshow(pic_1, cmap='gray')
            plt.title(msg)
            plt.xticks([])
            plt.yticks([])
            plt.subplot(122)
            plt.imshow(pic_2, cmap='gray')
            plt.title(msg_2)
            plt.xticks([])
            plt.yticks([])
    else:
        plt.subplot(131)
        plt.imshow(pic_1, cmap='gray')
        plt.title(msg)
        plt.xticks([])
        plt.yticks([])
        plt.subplot(132)
        plt.imshow(pic_2, cmap='gray')
        plt.title(msg_2)
        plt.xticks([])
        plt.yticks([])
        plt.subplot(133)
        plt.imshow(pic_3, cmap='gray')
        plt.title(msg_3)
        plt.xticks([])
        plt.yticks([])
    plt.savefig('show_out_test.png', dpi=500)
```

```python
def gaussian(pic):
    row, col = pic.shape
    x = np.arange(1, 256)
    y = np.arange(1, 256)
    for i in range(255):
        y[i] = 0
    for i in range(row):
        for j in range(col):
            y[pic[i][j] - 1] += 1

    noise = np.random.random((pic.shape))

    plt.subplot(221)
    plt.imshow(pic, cmap='gray')
    plt.xticks([])
    plt.yticks([])

    plt.subplot(223)
    plt.plot(x, y)
    # plt.ylim(0, 300)
    plt.xlim(0, 255)
```

```python
cr = readGRAY('data/01_cr.png')
gaussian(cr)
(cr.shape)
```

```python
from ipywidgets import interact
import ipywidgets as widgets



cr = readGRAY('data/01_cr.png')
row, col = cr.shape
sli = widgets.IntSlider(value=400, min=0, max=row)
def test():
    img = cr.copy()
    pos = sli.value
    x = range(0, col)
    y = cr[pos].copy()
    for i in range(col):
        for j in range(-5, 5):
            img[pos + j][i] = 255
    plt.subplot(211)
    plt.plot(x, y)
    plt.title("Gray Value")

    plt.subplot(212)
    plt.imshow(cr, cmap='gray')
    plt.xticks([])
    plt.yticks([])
    plt.title("Selected Line")
sli.observe(test)
sli
```
