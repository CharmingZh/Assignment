## 使用方法

要使用`showtext`创建一个图形，只需做以下工作。

- **加载字体**
- 打开图形设备(Open the graphics device)
- **声称你想使用`showtext`来渲染文字**
- 绘制
- 关闭设备

只有**加粗**的步骤是新的。
- 如果你想在全局范围内使用`showtext`，你可以调用函数`showtext_auto()`一次，之后所有的设备都会自动使用`showtext`来渲染;
- 如果你想更精细地控制代码的哪个部分应该使用`showtext`，可以使用函数`showtext_begin()`和`showtext_end()`。

```r
library(showtext)
#font_files()
showtext_auto(enable = TRUE)
font_add('Songti', 'Songti.ttc')

data <- read.table("Lecture 2/data/data.txt", header = T)
barplot(apply(data,1,mean))
```
